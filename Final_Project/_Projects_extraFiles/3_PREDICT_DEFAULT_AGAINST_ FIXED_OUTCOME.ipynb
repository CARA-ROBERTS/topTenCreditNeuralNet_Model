```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
cc_df = pd.read_csv('Credit_Data_Original.csv')
cc_df.head(1000)
```

```python
 # Assign X (data) and y (target)
X = cc_df.drop("DEFAULT", axis=1)
y = cc_df["DEFAULT"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:20]}")
print(f"First 10 Actual labels: {y_test[:20].tolist()}")
```

```python
#  Adjust dataFrame size to fit pediction default
new_df= cc_df.iloc[0:250]
new_df
```
